```python
from shapely.geometry import Polygon, LineString, Point
```

```python
p1 = Point(85,30)
```

```python
p1
```

```python
print(p1)
```

```python
p2 = Point(84,31)
l1 = LineString([p1,p2])
print(l1)
l1
```

```python
poly1 = Polygon([[82,21],[83,22],[82,23]])
print(poly1)
poly1
```

```python
df_shapely = gpd.GeoSeries([p1,p2,l1,poly1])
```

```python
df_shapely
```

```python
geoseries_shapely = gpd.GeoSeries([p1,p2,l1,poly1])
df_shapely = gpd.GeoDataFrame(geoseries_shapely)
df_shapely
```

```python
df_shapely.columns = ['geometry']
df_shapely.crs = 'EPSG:4326'
df_shapely
```

```python
df_shapely.plot()
```

```python
df_shapely.to_file(r'/content/output/shapely.geojson', driver='GeoJSON')
```
