```python
in_img = r'/content/data/srtm_57_08_1.tif'
ds = rasterio.open(in_img)
ds
```

```python
ds = rasterio.open(in_img)
data = ds.read()
data
```

```python
from itertools import count
with rasterio.open(r'/content/output/strm_out.tif', 'w',
                   driver = ds.driver,
                   height = ds.height,
                   width = ds.width,
                   count = ds.count,
                   crs = ds.crs,
                   transform = ds.transform,
                   dtype = data.dtype
                    ) as dst:
                     dst.write(data)
```
